```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)

# Path to CSV
csv_path = "../data/flights_sample_3m.csv"

# Read the CSV into a DataFrame
df = pd.read_csv(csv_path)

# Show the first few rows
df.head()
```

```python
# Show the column names
print(df.columns)
```

```python
df.describe(include='all')
```

Looking at unique values, probably can't use One Hot Encoding for AIRLINE, ORIGIN, and DEST given the high cardinality. Need someway to encode these categorical variables though.

```python
# Look at the data types of the columns
df.info()
```

Important variables like CANCELLED, CRS_DEP_TIME, CRS_ARR_TIME are already numerical, which means we likely don't have to do any work on them.

```python
# Look at the cancellation rate
df["CANCELLED"].value_counts(normalize=True)
```

```python
# Look at what columns have the greatest number of missing values
df.isnull().mean().sort_values(ascending=False)
```

```python
# Cancellation rate by airline
df.groupby("AIRLINE")["CANCELLED"].mean().sort_values()
```

```python
# Cancellation rate by origin, descending order
(
    df.groupby("ORIGIN")
      .agg(rate=("CANCELLED", "mean"), count=("CANCELLED", "size"))
      .query("count > 10000")
      .sort_values("rate", ascending=False)
)
```

Rather significant difference in the cancellation rates, around 4% for the highest and around 1.2% lowest.

```python
# Cancellation rate by destination, descending order
(
    df.groupby("DEST")
      .agg(rate=("CANCELLED", "mean"), count=("CANCELLED", "size"))
      .query("count > 10000")
      .sort_values("rate", ascending=False)
)
```

```python
# Plot of cancellation rate by departure hour
df["dep_hour"] = df["CRS_DEP_TIME"] // 100
df.groupby("dep_hour")["CANCELLED"].mean().plot(kind="bar")
```

Somewhat of a pattern is observed, where 11am seems to have lower cancellation rates, having decreased from 5am, and increasing till 10pm.

```python
# Cancellation rate by distance (grouped into 10 bins)
df.groupby(pd.cut(df["DISTANCE"], bins=10))["CANCELLED"].mean()
```

Seems like longer flights are less likely to be cancelled

```python
corr_matrix = df.corr(numeric_only=True)

cancelled_corr = corr_matrix["CANCELLED"].sort_values(ascending=False)

plt.figure(figsize=(4, 12))
# Convert Series to DataFrame for heatmap (needs 2D data)
sns.heatmap(cancelled_corr.to_frame(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation of CANCELLED with other variables")
plt.show()
```

Correlation matrix suggests that these linear correlations are weak since all have values close to 0. Cancellation patterns are likely non-linear/categorical.
